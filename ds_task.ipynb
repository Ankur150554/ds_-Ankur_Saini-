```python
merged_df = pd.merge(df,df1 , on='date', how='left')
```

```python
merged_df.head()
```

We create:

* **abs_pnl**: absolute profit/loss per trade

* **is_profit**: whether the trade was profitable

* **usd_volume**: USD notional of each trade (already given as Size USD)

```python

merged_df['Closed PnL'] = pd.to_numeric(merged_df['Closed PnL'], errors='coerce')

merged_df['abs_pnl'] = merged_df['Closed PnL'].abs()
merged_df['is_profit'] = merged_df['Closed PnL'] > 0
merged_df['usd_volume'] = pd.to_numeric(merged_df['Size USD'], errors='coerce')


merged_df[['Closed PnL', 'abs_pnl', 'is_profit', 'usd_volume']].head()
```

```python
avg_pnl_by_sentiment = merged_df.groupby('classification')['Closed PnL'].mean()
print("Average Profit/Loss by Market Sentiment:")
print(avg_pnl_by_sentiment)
```

```python
avg_pnl_by_sentiment.plot(kind='bar', color='orange')
plt.title('Average Trader Profit/Loss by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Closed PnL')
plt.grid(True)
plt.show()
```

```python
# Group by sentiment and compute key metrics
group_cols = ['classification']

agg_by_sent = merged_df.groupby(group_cols).agg(
    trades=('Account', 'count'),
    total_usd_volume=('usd_volume', 'sum'),
    avg_usd_volume=('usd_volume', 'mean'),
    total_pnl=('Closed PnL', 'sum'),
    avg_pnl_per_trade=('Closed PnL', 'mean'),
    median_pnl=('Closed PnL', 'median'),
    win_rate=('is_profit', 'mean'),
    avg_fee=('Fee', 'mean')
).reset_index()

# Win rate as percentage
agg_by_sent['win_rate_pct'] = agg_by_sent['win_rate'] * 100

agg_by_sent.sort_values('trades', ascending=False)
```

```python
# Plot win rate by sentiment
plt.figure()
plt.bar(agg_by_sent['classification'], agg_by_sent['win_rate_pct'])
plt.title("Win Rate (Profit) by Market Sentiment")
plt.xlabel("Sentiment Class")
plt.ylabel("Win Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```



* Win rate varies by market sentiment, with **Extreme Greed** having the highest success rate.
* **Extreme Fear** shows the lowest win rate among all sentiment classes.
* Neutral, Fear, and Greed sentiments sit in the mid-range with similar win-rate performance.


```python
risk_by_sentiment = merged_df.groupby('classification')['Closed PnL'].std()
print("Risk (PnL Std Dev) by Market Sentiment:")
print(risk_by_sentiment)

risk_by_sentiment.plot(kind='bar', color='red')
plt.title('Risk Level by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('PnL Standard Deviation')
plt.grid(True)
plt.show()
```



* Market risk is highest during **Extreme Fear and Greed**, where PnL fluctuations become large and unpredictable.
* Risk decreases during **Fear and Extreme Greed**, showing more controlled price movement compared to the extremes.
* **Neutral sentiment is the safest phase**, with the lowest volatility and most stable trading conditions.


```python
volume_by_sentiment = merged_df.groupby('classification')['Size USD'].sum()
print("\nTotal Trade Volume by Market Sentiment:")
print(volume_by_sentiment)

volume_by_sentiment.plot(kind='bar', color='blue')
plt.title('Trade Volume by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Total Trade Size')
plt.grid(True)
plt.show()
     
```

```python
position_by_sentiment = merged_df.groupby('classification')['Start Position'].mean()
print("\nAverage Start Position by Market Sentiment:")
print(position_by_sentiment)

position_by_sentiment.plot(kind='bar', color='purple')
plt.title('Average Start Position by Market Sentiment')
plt.xlabel('Market Sentiment')
plt.ylabel('Average Start Position')
plt.grid(True)
plt.show()
     
```



* Average start positions vary strongly across market sentiment categories.
* Greed sentiment shows a very large negative start position, while Extreme Greed shows a high positive one.
* Extreme Fear, Fear, and Neutral sentiments have comparatively small average start positions.


```python
merged_df['usd_volume'] = pd.to_numeric(merged_df['Size USD'], errors='coerce')
merged_df['trade_datetime'] = pd.to_datetime(
    merged_df['Timestamp IST'],
    format='%d-%m-%Y %H:%M',
    errors='coerce'
)
merged_df['trade_date'] = merged_df['trade_datetime'].dt.date

daily_stats = merged_df.groupby('trade_date').agg(
    daily_trades=('Account', 'count'),
    daily_usd_volume=('usd_volume', 'sum'),
    daily_pnl=('Closed PnL', 'sum')
).reset_index()

daily_stats = pd.merge(
    daily_stats,
    merged_df[['date', 'value', 'classification']],
    left_on='trade_date',
    right_on='date',
    how='left'
)
daily_stats.head()
```

```python
plt.figure(figsize=(10, 5))

plt.plot(
    daily_stats['trade_date'],
    daily_stats['daily_usd_volume'],
    linestyle='--'
)   # removed sns, default Matplotlib color will be used

plt.title("Daily USD Volume Over Time")
plt.xlabel("Trading Date")
plt.ylabel("Values")
plt.xticks(rotation=45)

plt.tight_layout()   # avoid cutting labels

plt.show()
```



* Daily USD trading volume stayed very low and stable for most of the early period.
* Starting around late 2024 and especially in early 2025, volume increased sharply with large spikes.
* The graph shows rapid volatility and significantly higher market activity toward the end of the timeline.


```python
plt.figure(figsize=(8, 6))

# Get unique sentiment classes and assign a color to each
sentiments = daily_stats['classification'].unique()
colors = plt.cm.tab10(range(len(sentiments)))  # automatic color palette

for sentiment, color in zip(sentiments, colors):
    subset = daily_stats[daily_stats['classification'] == sentiment]
    plt.scatter(
        subset['daily_usd_volume'],
        subset['daily_pnl'],
        label=sentiment,
        color=color
    )

plt.xlabel("USD Volumes per Day")
plt.ylabel("Profit and Loss per Day")
plt.title("Volume vs PnL by Sentiment")
plt.legend(title="Sentiment Class")

plt.tight_layout()
p
plt.show()
```
